Generate Keywords from browser history

```python
import pandas as pd
from keybert import KeyBERT
from google.colab import files
from keybert import KeyBERT
import pandas as pd 
import neattext.functions as nfx

# Read csv file from drive
df = pd.read_csv( '/content/drive/MyDrive/Final year project/history (1).csv' )
# Remove stopwords from title
df['clean_title'] = df['title'].apply(nfx.remove_stopwords)
# loop x for every row in the csv file
for x in range(1016) :
  doc = df['clean_title'][x]
  sep = '-'
  # strip everything after '-' in the title
  stripped = doc.split(sep, 1)[0]
  kw_model = KeyBERT()
  # Generate keywords
  key = kw_model.extract_keywords( stripped , keyphrase_ngram_range=(1, 3) , top_n=1 )
  print ( "Title : " , df['title'][x] , "    Keyword : " , key )
```
